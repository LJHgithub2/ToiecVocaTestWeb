```python
pdf_file = "/content/drive/Othercomputers/내 노트북/code/ToiecVocaTestWeb/Backend/data/TK/TK_voca.pdf"
output_images_dir = "/content/drive/Othercomputers/내 노트북/code/ToiecVocaTestWeb/Backend/data/TK/images"
output_text_dir = "/content/drive/Othercomputers/내 노트북/code/ToiecVocaTestWeb/Backend/data/TK/extracted_data"
```

```python
from pdf2image import convert_from_path
from PIL import Image
import pytesseract
import os

# PDF를 고해상도로 이미지로 변환
images = convert_from_path(pdf_file, dpi=300)

# 변환된 이미지를 저장하고, 잘라서 OCR로 텍스트 추출
for i, image in enumerate(images[:2]):
    # 이미지 저장 경로 설정
    original_image_path = os.path.join(output_images_dir, f'page_{i+1}.png')

    # 원본 이미지 저장
    image.save(original_image_path, 'PNG')

    # 이미지 크기 얻기
    width, height = image.size

    # 원본 파일 OCR 텍스트 추출
    # text = pytesseract.image_to_string(image, lang='eng+kor')
    # with open(os.path.join(output_text_dir, f'page_{i+1}.txt'), 'w', encoding='utf-8') as f:
    #     f.write(text)

    # print(f"Extracted text from images: {output_text_dir}")

    # 왼쪽 반 이미지 자르기
    left_image = image.crop((0, 0, width // 2, height))
    left_image_path = os.path.join(output_images_dir, f'page_{i+1}_left.png')
    left_image.save(left_image_path, 'PNG')

    # 오른쪽 반 이미지 자르기
    right_image = image.crop((width // 2, 0, width, height))
    right_image_path = os.path.join(output_images_dir, f'page_{i+1}_right.png')
    right_image.save(right_image_path, 'PNG')

    # OCR로 텍스트 추출
    left_text = pytesseract.image_to_string(left_image, lang='eng+kor')
    right_text = pytesseract.image_to_string(right_image, lang='eng+kor')

    # 텍스트를 파일로 저장
    with open(os.path.join(output_text_dir, f'page_{i+1}_left.txt'), 'w', encoding='utf-8') as f:
        f.write(left_text)

    with open(os.path.join(output_text_dir, f'page_{i+1}_right.txt'), 'w', encoding='utf-8') as f:
        f.write(right_text)

    print(f"Extracted text from images: {left_image_path} and {right_image_path}")
```

```python
def merge_strings(lst):
    result = []
    temp = ""

    for item in lst:
        if item.strip() == "":  # 공백 문자열일 경우
            if temp:  # 임시 문자열이 비어있지 않으면
                result.append(temp.strip())  # 임시 문자열을 결과 리스트에 추가
                temp = ""  # 임시 문자열 초기화
            result.append(item)  # 공백 문자열 자체를 결과 리스트에 추가
        else:
            temp += item + " "  # 임시 문자열에 현재 문자열과 공백을 추가

    if temp:  # 마지막 임시 문자열이 비어있지 않으면
        result.append(temp.strip())  # 결과 리스트에 추가

    return result

# 테스트
lst = ["hi", "     " ,"myname","jonghyeon","yes","        ","jong","ya","        ","adfasf"]
print(merge_strings(lst))
def text_to_list(text):
    return text.split('\n')

# 예시 텍스트
example_text = """첫 번째 줄
두 번째 줄
세 번째 줄"""

# 텍스트를 리스트로 변환
text_list = text_to_list(example_text)

# 결과 출력
print(text_list)
```

```python
from pdf2image import convert_from_path
import os

def capture_slide_from_pdf(pdf_path, slide_number, output_dir):
    # PDF 파일에서 이미지로 변환
    images = convert_from_path(pdf_path, first_page=slide_number, last_page=slide_number)

    if images:
        image_path = os.path.join(output_dir, f"slide_{slide_number}.png")
        images[0].save(image_path, "PNG")
        print(f"Saved slide {slide_number} as image: {image_path}")
    else:
        print(f"No image captured from slide {slide_number}")

# PDF 파일 경로, 캡처할 슬라이드 번호, 이미지를 저장할 디렉토리 경로 지정
pdf_path = "/content/자료구조.pdf"  # PDF 파일 경로를 지정하세요
slide_number = 1  # 캡처할 슬라이드 번호를 지정하세요
output_dir = "/content"  # 이미지를 저장할 디렉토리 경로를 지정하세요

capture_slide_from_pdf(pdf_path, slide_number, output_dir)
```

```python
import re
import pytesseract
from PIL import Image

def ocr_image_to_text(image_path):
    # 이미지를 불러옵니다.
    img = Image.open(image_path)

    # 이미지를 텍스트로 변환합니다.
    text = pytesseract.image_to_string(img)

    return text

def process_text(text):
    # 숫자와 마침표, 그리고 공백으로 이루어진 패턴을 찾아 제거합니다.
    processed_text = re.sub(r'\d+\. ', '', text)
    processed_text = re.sub(r'\. ', '', processed_text)  # '.'만 있는 경우에도 제거합니다.

    return processed_text

# 이미지 파일 경로를 지정합니다.
image_path = '/content/slide_1.png'

# 이미지에서 텍스트 추출
text = ocr_image_to_text(image_path)

# 텍스트 처리
processed_text = process_text(text)

print("이미지에서 추출한 텍스트:")
print(processed_text)
```
